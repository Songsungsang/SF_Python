```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

x = np.arange(5)

plt.bar(x, data)
plt.xticks(x, categories, rotation=45)
plt.yticks(np.arange(0,51,10))
plt.grid(True, "both")

plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

#더 쉽고 간편하게 x,y 짜기 가능
plt.bar(categories, data)
plt.xticks(rotation=45)
plt.ylim(0,50)
plt.grid(True, "both")

plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```
