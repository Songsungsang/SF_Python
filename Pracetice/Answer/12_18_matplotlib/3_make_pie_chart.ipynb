```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
size = [15, 30, 45, 10]
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
explode = (0, 0.1, 0, 0)

plt.pie(size, labels=labels, autopct = '%.1f%% ', explode=explode, startangle=90, shadow=True) # 90도 돌리고 그림자 넣고

plt.show()
```
