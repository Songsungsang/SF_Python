```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 0 10  100 

# sin cos   
y_sin = np.sin(x)
y_cos = np.cos(x)

line1 = plt.plot(x, y_sin, label='sin(x)', color = 'b') # x, y, , 
line2 = plt.plot(x, y_cos, label='cos(x)', color = 'r')

plt.legend(labels= ("'sin(x)", "cos(x)")) #   
plt.grid(True, "both")
plt.title("Sine and Cosine Functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
