```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 0에서 10 사이를 100개로 나눈거

# sin과 cos 함수의 값을 계산
y_sin = np.sin(x)
y_cos = np.cos(x)

line1 = plt.plot(x, y_sin, label='sin(x)', color = 'b') # x, y, 레이블, 색깔
line2 = plt.plot(x, y_cos, label='cos(x)', color = 'r')

plt.legend(labels= ("'sin(x)", "cos(x)")) # 레전드에 레이블들 표시
plt.grid(True, "both")
plt.title("Sine and Cosine Functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
