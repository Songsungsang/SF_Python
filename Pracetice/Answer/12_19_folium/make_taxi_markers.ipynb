```python
import pandas as pd
import folium
```

```python
m = folium.Map(location = [37.547631, 126.942463], zoom_start = 14,
               width = 800, height = 450) # m 초기 설정 및 필요에 따라 초기화 할떄도 이 블럭 사용
```

```python
taxi_path = "../../dataset/택시승차대 현황.csv"
taxi_info = pd.read_csv(taxi_path, encoding='cp949')
```

```python
taxi = taxi_info.loc[:, ["위치명", "위도", "경도"]]
for index, row in taxi.iterrows():
    folium.Marker([row["위도"], row["경도"]], # 각 행의 위도와 경도의 위치에
            icon=folium.Icon(color="lightblue", icon='fa-solid fa-taxi', prefix='fa-solid')).add_to(m) #마커 표시 # add_to로 지도에 추가 필수
    folium.Circle([row["위도"], row["경도"]], 
                  radius = 300, color = '#ffffff', fill_color = '#00ff00').add_to(m) # 이정도 크기에 이런 색깔의 선과 채우기 색깔로
```
