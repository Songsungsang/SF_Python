```python
import requests
from bs4 import BeautifulSoup
```

```python
search_input = input("검색어를 입력해주세요 :")
response = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=" + search_input) # 이 페이지로 보내는 리퀘스트 결과 받기
html = response.text # 리퀘스트를 텍스트로 받음
soup = BeautifulSoup(html, 'html.parser')
```

```python
news_ul = soup.select_one('ul.list_news') #네이버 검색에서 뉴스가 모여있는 ul 클래스에 까지 접촉
print(news_ul)
```

```python
news_li = news_ul.select('ul.list_news > li') # ul 안에 뉴스가 담겨져 있는 li까지 접촉
print(search_input)
for sub_li in news_li: # 각 li 마다
    news_each = sub_li.find('a', class_='news_tit') # 가장 큰 뉴스 타이틀을 가진 a 찾기
    news_detail = news_each['title'] + " " + news_each['href'] # 내용 넣기
    print("=" * 30 ,"\n",news_detail) # 내용 출력
    sub_news = sub_li.findAll('a', class_="elss sub_tit") # 하위 뉴스 찾기
    for i in sub_news: # 각 하위 뉴스마다
        sub_news_detail = " " + i['title'] + " " + i['href'] # 내용 넣기
        print(sub_news_detail) # 내용 출력
```
